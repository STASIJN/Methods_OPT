```python
from matplotlib import pyplot as plt
import numpy as np
from math import *


```

```python
def func(x):
    return 10*x*np.log(x) - x**2/2

left = 0.1
right = 1
arr = np.linspace(left, right, 100)
plt.plot(arr,func(arr))
```

```python
def OverValues(a, b, e, func):
    count_iter = 0

    x_min = a
    a += e
    while(func(x_min) > func(a) and a <= b):
        count_iter +=2
        plt.scatter(x_min, func(x_min), color='green')
        x_min = a
        a += e    
    return x_min, func(x_min), count_iter

```

```python
for i in (0.01, 0.001, 0.0001):
    plt.subplot(1, 3, 1 if i == 0.01 else 2 if i == 0.001 else 3)
    plt.plot(arr,func(arr))
    X_min, Y_min, count = OverValues(left, right, i, func)
    plt.scatter(X_min, func(X_min), color='red')
    print(X_min, Y_min, count)



plt.show()
```

```python
def BitwiseSearch(a, b, e, func):
    delta = e*100
    count_iter = 0
    while a + delta + delta >= b:
        delta -= 1

    x_min_2 = a
    x_min_1 = x_min_2
    while(abs(delta) > e):
        x_min_1 = x_min_2
        x_min_2 += delta
        plt.scatter(x_min_2, func(x_min_2), color='green')
        while(func(x_min_2) <= func(x_min_1) and a<= x_min_2 <= b):
            x_min_1 = x_min_2
            x_min_2 += delta
            count_iter +=3
        delta /= -4
        plt.scatter(x_min_2, func(x_min_2), color='blue')
    return x_min_1, func(x_min_1), count_iter

```

```python
for i in (0.01, 0.001, 0.0001):
    plt.subplot(1, 3, 1 if i == 0.01 else 2 if i == 0.001 else 3)
    plt.plot(arr,func(arr))
    X_min, Y_min, count = BitwiseSearch(left, right, i, func)
    plt.scatter(X_min, func(X_min), color='red')
    print(X_min, Y_min, count)



plt.show()
```

```python
def Dichotomy(a, b, e, func):
    delta = e
    count_iter = 0
    plt.scatter(a, func(a), color='blue')
    plt.scatter(b, func(b), color='green')
    while (b-a)/2 > e:
        x_1 = (b + a - delta)/2 
        x_2 = (b + a + delta)/2
        count_iter += 2 
        if func(x_2) >= func(x_1):
            b = x_2
            plt.scatter(b, func(b), color='green')
        else:
            a = x_1
            plt.scatter(a, func(a), color='blue')   
    return (b + a)/2, func((b + a)/2), count_iter

    
```

```python
def Half(a, b, e, func):
    delta = e/10
    count_iter = 0
    plt.scatter(a, func(a), color='blue')
    plt.scatter(b, func(b), color='green')
    while (b-a)/2 > e:
        x_2 = (a + (b-a)/2) + delta
        x_1 = (a + (b-a)/2) - delta
        count_iter += 2 
        if func(x_2) >= func(x_1):
            b = x_2
            plt.scatter(b, func(b), color='green')
        else:
            a = x_1
            plt.scatter(a, func(a), color='blue')   
    return (b + a)/2, func((b + a)/2), count_iter
```

```python
def Golden(a, b, e, func):
    delta = (sqrt(5) - 1)/2
    count_iter = 0
    plt.scatter(a, func(a), color='blue')
    plt.scatter(b, func(b), color='green')
    while (b-a)/2 > e:
        x_2 = a+delta*(b-a)
        x_1 = a + b - x_2
        count_iter += 2 
        if func(x_2) >= func(x_1):
            b = x_2
            plt.scatter(b, func(b), color='green')
        else:
            a = x_1
            plt.scatter(a, func(a), color='blue')   
    return (b + a)/2, func((b + a)/2), count_iter

```

```python
def Parabol(a, b, e, func):
    x = (b+a)/2
    count_iter = 0

    plt.scatter(a, func(a), color='green')
    plt.scatter(x, func(x), color='green')
    plt.scatter(b, func(b), color='green')
    
    kof_a = (func(x) - func(a))/(x - a)
    kof_b = 1/(b-x)*( (func(b)-func(a))/(b-a) - (func(x) - func(a))/(x-a) )
    count_iter += 3 
    x_min = 1/2*(a + b - kof_a/kof_b)

    if x_min > x:
        b = x_min
    else:
        a = x_min

    x_min_tmp = x_min
    x_min = 2*e+x_min_tmp

    while(abs(x_min - x_min_tmp) > e):
        x_min = x_min_tmp
        kof_a = (func(x) - func(a))/(x - a)
        kof_b = 1/(b-x)*( (func(b)-func(a))/(b-a) - (func(x) - func(a))/(x-a) ) 
        count_iter += 3 
        x_min_tmp = 1/2*(a + b - kof_a/kof_b)

        if x_min_tmp > x:
            b = x_min_tmp
        else:
            b = x
            x = x_min_tmp
        
        plt.scatter(a, func(a), color='blue')
        plt.scatter(x, func(x), color='blue')
        plt.scatter(b, func(b), color='blue')

        x_min = x_min_tmp
        kof_a = (func(x) - func(a))/(x - a)
        kof_b = 1/(b-x)*( (func(b)-func(a))/(b-a) - (func(x) - func(a))/(x-a) ) 
        count_iter += 3 
        x_min_tmp = 1/2*(a + b - kof_a/kof_b)

        if x_min_tmp > x:
            a = x
            x = x_min_tmp
        else:
            a = x_min_tmp
        
        plt.scatter(a, func(a), color='green')
        plt.scatter(x, func(x), color='green')
        plt.scatter(b, func(b), color='green')
    return x_min, func(x_min), count_iter





```

```python
for i in (0.01, 0.001, 0.0001):
    plt.subplot(1, 3, 1 if i == 0.01 else 2 if i == 0.001 else 3)
    plt.plot(arr,func(arr))
    X_min, Y_min, count = Parabol(left, right, i, func)
    plt.scatter(X_min, func(X_min), color='red')
    print(X_min, Y_min, count)



plt.show()
```
